```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('/content/laptop_prices.csv')


```

```python


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Price_euros'], bins=20, color='skyblue')
plt.xlabel('Price (Euros)')
plt.ylabel('Frequency')
plt.title('Histogram of Laptop Prices')

plt.subplot(1, 2, 2)
plt.hist(df['Ram'], bins=10, color='salmon')
plt.xlabel('RAM (GB)')
plt.ylabel('Frequency')
plt.title('Histogram of RAM Sizes')

plt.tight_layout()
plt.show()
```

```python

type_counts = df['TypeName'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Laptop Types')
plt.axis('equal')
```

```python


brand_counts = df['Company'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(brand_counts.index, brand_counts.values, color='skyblue')
plt.xlabel('Laptop Brand')
plt.ylabel('Number of Laptops')
plt.title('Distribution of Laptop Brands')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python


os_counts = df['OS'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(os_counts.index, os_counts.values, color='lightgreen')
plt.xlabel('Operating System')
plt.ylabel('Number of Laptops')
plt.title('Distribution of Laptop Operating Systems')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python


plt.figure(figsize=(12, 6))
sns.countplot(x='Company', data=df)
plt.xlabel('Laptop Brand')
plt.ylabel('Number of Laptops')
plt.title('Number of Laptops per Brand')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(df['Ram'], df['Price_euros'], alpha=0.5)
plt.xlabel('RAM (GB)')
plt.ylabel('Price (Euros)')
plt.title('Laptop Price vs. RAM Size')
plt.show()

```
